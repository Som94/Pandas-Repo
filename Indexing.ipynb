# Copy one DataFrame to Another

```python
# By Using copy() method

df2=df.copy()
```

```python
df2
```

```python
#By using = opretor, but there is oner problem with '=' oparetor , when perform any modification on copied DataFrame ,It will be
# affecterd to our original DataFrame

df3=df2
```

```python
df3
```

```python
df3.set_index('EST')   # It will not  set index / change index permanently
```

```python
df3
```

```python
# In above we see that df3.set_index('EST')  ,It will not  set index / change index permanently.

# If we want to  set index / change index permanently ,then we specify inplace attribute is True (inplace=True)

#Syntax : <DataFrame name>.set_index( '<column name>', inplace=Ture)

# But there is another way without using inplace attribute, just we can store set indexing value into varialbe and call that
```

```python
df3.set_index('EST', inplace=True)
```

```python
df3
```


We can specify the index column when we create the dataframe , for this we have to specify 'index_col' attribute

syntax:
    
    <Data Frame name>=pd.<read which type of file>('<File name>', index_col=<specify column name>)
    
ex:

    df=pd.read_csv("E:\Aroha Tech\Pandas Session\FILE.csv", index_col='EST')
    

But recomanded to go for only unique column value to set as index column

```python
df4=pd.read_csv("E:\Aroha Tech\Pandas Session\FILE.csv", index_col='EST')
```

```python
df4
```

# sort data

```python
# We can sort data by index wise, we can use sort_index()

#syntax :  <Data Frame Name>.sort_index()     (By default it return ascending order sorting value)

#            <Data Frame Name>.sort_index(ascending=False)      (By default it return descending order sorting value)
```

```python
df4.sort_index()     # (By default it return ascending order sorting value)
```

```python
df4.sort_index(ascending=False)      # (By default it return descending order sorting value)
```
