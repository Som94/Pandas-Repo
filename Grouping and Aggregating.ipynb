```python
filt=survey_df['Country']=='India'
survey_df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
# survey_df['LanguageWorkedWith']
```

```python
# Display the countywise how many people knew the python
```

```python
country_grp= survey_df.groupby(['Country'])
country_grp['LanguageWorkedWith'].str.contains('Python').sum()  # 'SeriesGroupBy' object has no attribute 'str', 
                                                                    # So, we go for apply() method
```

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
# Display the countywise how many % people knew the python

# country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
# for this we have to first count total no. of repondents for each country
country_respondents=survey_df['Country'].value_counts()
country_respondents
```

```python
# Then now count total number of people who knew python

country_uses_pyhton= country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())

country_uses_pyhton
```

```python
# Now we combine tese two by Using concat() method

python_df=pd.concat([country_respondents,country_uses_pyhton], axis='columns', sort=False)
```

```python
python_df
```

```python
python_df.rename(columns={'Country':'NumRespondents','LanguageWorkedWith':'NumknowsPython'}, inplace=True)
```

```python
python_df
```

```python
# so finally here we calculate how many percentage people know python

#       percentage=part/whole * 100

python_df['PPKnowsPython']=(python_df['NumknowsPython']/python_df['NumRespondents']) * 100
```

```python
python_df
```

```python
# sort the country by the largest percentage of people who know python

python_df.sort_values(by='PPKnowsPython', ascending=False, inplace=True)
```
