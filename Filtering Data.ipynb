# And , Or oparetor in pandas 

We can't use in pandas our python's and , or oparetor , instead of that we can use some special oparetors

& ---> for and 

| ---> for or

```python
# Q. Display the email id where first name is 'John' and last name is 'Doe'
```

```python
filt=(df1['first']=='John') & (df1['last']=='Doe')   # When we use '&' or '|' , make sure we have to keep each part inside '()' ,
                                                    # Other wise we will get Type Error
df1.loc[filt,'email']
```

```python
# Q. Display the email id where first name is 'Coery' or last name is 'Doe'
```

```python
filt=(df1['first']=='Coery') | (df1['last']=='Doe')

df1.loc[filt,'email']
```

Negative filter

```python
# For negetive filter , just we add prefix '-' (minus) to Series object of boolean value.
```

```python
# Q. Display the email id where first name is not 'John' or last name is not'Schafer' 
```

```python
filt=(df1['first']=='John') | (df1['last']=='Schafer')
```

```python
df1.loc[filt,'email']     # This is not our acctual data according to th question.
```

```python
# To get actual data according to question we have to add '-'' prefix to filt.

df1.loc[-filt,'email']
```
