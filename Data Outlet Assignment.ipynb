# Convert item fat content  intems into only Low Fat and Regular


```python
convrt=df['Item_Fat_Content']                   
```

```python
Item_Fat_Content_dup=[]
```

```python
for i in convrt:
    if i[0].upper()=='L':
        i='Low Fat'
        #print(i)
        Item_Fat_Content_dup.append(i)
    else:
        i='Regular'
        #print(i)
        Item_Fat_Content_dup.append(i)
```

```python
Item_Fat_Content_dup
```

```python
df['Item_Fat_Content']=Item_Fat_Content_dup
```
