```python
import pandas as pd
```

```python
people={'first':['Coery','Jane','John','Adam'],
      'last':['Schafer','Doe','Doe','Doe'],
       'email':['CoeryMSchafer@gmail.com','JaneDoe@gmail.com','JohnDoe@gmail.com','A@email.com']
     }
```

```python
df=pd.DataFrame(people)
```

```python
df
```

# sort_values()

```python
#
```

```python
#Q. sort by last name column
df.sort_values(by='last')
```

```python
#If we want to sort values by descending order, then just 'specify ascending=False'
```

```python
df.sort_values(by='last', ascending=False)
```

```python
# We can sort multiple columns , for this we must pass list of multple columns.
# But sort will happen according to by which order we pass columns
```

```python
df.sort_values(by=['last','first'])  # In this case 1st last column sort then last
```

```python
df.sort_values(by=['last','first'], ascending=False)
```

```python
df.sort_values(by=['last','first'], ascending=[False, True])  #It will sort last column as descending order then first column as
                                                                        #  ascending order
```

```python
df.sort_values(by=['last','first'], ascending=[False, True], inplace=True) 
```

```python
df
```

```python
# Sort a single column
df['last'].sort_values()
```

# nlargest()

```python
#Q. Find the 10 largest salary
```

```python
df1['ConvertedComp'].nlargest(10)    # It will return Series object of  10 highest salry
```

```python
# If we want display 10 highest salary along with the all details, then

# Syntax : <DataFrame Name>.nlargest(<no. of value> , '[<Column name>]')
```

```python
df1.nlargest(10,['ConvertedComp'])
```

```python
#Q. Display 10 highest salary along with the 10 highest Age.

df1.nlargest(10,['ConvertedComp','Age'])
```
