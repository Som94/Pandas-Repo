```python
import  pandas as pd
import numpy as np
```

```python
people={'first':['Coery','Jane','John','Chris',np.nan,None, 'NA'],
      'last':['Schafer','Doe','Doe','Schafer',np.nan,np.nan,'Missing'],
       'email':['CoeryMSchafer@gmail.com','JaneDoe@gmail.com','JohnDoe@gmail.com',None,np.nan,'Anonymous@gmail.com','NA'],
        'age' : ['33','55','63','36',None,None,'Missing']
     }  
```

```python
df=pd.DataFrame(people)
```

```python
df
```

# dropna()

```python
# Syntax :
#          < Data Frame Name >.dropna( axis='<index/columns>' , how=' any / all', subset=['<Specific column name>'] )

# the axis attribute can set either index or column, theat is going to tell our pandas we drop our missing value in index or 
#  columns.
#  the how attribute by default take 'any' as default value. if we specify value as all, then if all values are missing in a row
#  then that row will be drop other wise not.

# dropna using in bacjkground some default arguments
```

```python
# drop the rowa which have missing first, last name column value

df.dropna()
```

```python
df.dropna(axis='index',how='any' )
```

```python
df.dropna(axis='index',how='all' )
```

```python
df.dropna(axis='columns',how='any' )    # If any column contains missing/nan value then it will be drop
```

```python
# Drop a specific row which contain a missing value

#drop a row which missing email address

df.dropna(axis='index', how='any', subset=['email'])
```

```python

df.dropna(axis='index', how='all', subset=['last','email'])   # if any fields last and email column's corressponding row value
                                                              # present , then it will not drop
```

```python
# convert NA and missing values into NaN 

df.replace('NA',np.nan, inplace=True)
df.replace('Missing',np.nan, inplace=True)
```

```python
df
```

```python

df.dropna(axis='index', how='all', subset=['last','email'])
```
