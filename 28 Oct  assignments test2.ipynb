1.Write a Pandas program to split a dataset, group by one column and get mean, min, and max values by group.
Using the following dataset find the mean, min, and max values of purchase amount (purch_amt) group by customer id (customer_id).


```python
group_by_customer_id=order_df.groupby('customer_id').agg({'purch_amt':['mean','min','max']})
```


4. Write a Pandas program to split the following dataframe into groups and calculate monthly purchase amount.


```python
order_df['ord_date']= pd.to_datetime(order_df['ord_date']) 
```

```python

monthly_purchase_amount = order_df.set_index('ord_date').groupby(pd.Grouper(freq='M')).agg({'purch_amt':sum})
```

```python
monthly_purchase_amount
```
