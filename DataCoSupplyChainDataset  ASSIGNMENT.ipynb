Q8. List all cities and respective no. of pending orders in India ?

```python
df8=df.copy()
```

```python
list1=df8.groupby(['Order Country','Order City'])['Order Status']
```

```python
list2=[]
for i in list1:
    if i[0][0]=='India':
        list2.append(i)
```

```python
result_list=[]
for j in list2:
    #print(j[0],j[1])
    for k in j[1]:
        if k=='PENDING':
           result_list.append((j[0],k)) 
    
```

```python
result_list
result_list_dict={}

for m in result_list:
    result_list_dict[m]=result_list_dict.get(m,0)+1
```

```python
result_list_dict
```
