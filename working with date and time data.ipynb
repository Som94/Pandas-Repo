```python

# If Date column already in the way it's formated like the pandas can parse them ,then we don't need to pass anithing
# But in case of above file Date column is not well formatting , so we ahve to pass a function that convert each string
# to date time object , so for that we use lambda function
```

```python
d_parser=lambda x : pd.datetime.strptime(x ,'%Y-%m-%d %I-%p')

df2=pd.read_csv(r'E:\Aroha Tech\Pandas Session\Pandas\ETH_1h.csv', parse_dates=['Date'], date_parser=d_parser)
```

```python
# To display all dates Day name , we have to access 'dt' class of day_name() method 
# 'dt' is the class of pandas series object
```
